AR(1), $\phi$=0.5
1. Time plot, check whether data is stationary or not.


```r
AR1<-arima.sim(list(ar=c(0.5)), n=200) #ARIMA simulation, coefficient of AR
ts.plot(AR1, main='AR(1)')
```

2. Choose the model what I have to use, considering ACF and PACF. (Ex. AR, MA, ARMA....etc)

```r
par(mfrow=c(1,2))
acf(AR1, main="ACF of AR(1)")
pacf(AR1, main='PACF of AR(1)')
```

AR(2), $\phi_{1}$=0.6, $\phi_{2}$=0.1

```r
par(mfrow=c(1,1))
AR2<-arima.sim(list(ar=c(0.6, 0.1)), n=200)
ts.plot(AR2, main='AR(2)')
```

```r
par(mfrow=c(1,2))
acf(AR2,main="AR(2) of ACE")
pacf(AR2,main="AR(2) of PACF")
```

How about AR(1) with $\phi$ = 0.9?
1. time plot

```r
AR1<-arima.sim(list(ar=c(0.9)), n=200) #ARIMA simulation, coefficient of AR
ts.plot(AR1, main='AR(1)')
```

2. check the ACF and PACF
As you can see the plots,the decreasing trend of ACF is weak. So, the data might be nonstationary because of $\phi$ almost equals to 1. It is difficult to check whether the data is stationary or not. So, we have to check it by using UNIT ROOT TEST.

```r
par(mfrow=c(2,1))
acf(AR1, main="ACF of AR(1)")
pacf(AR1, main='PACF of AR(1)')
```

MA(1), $Z_{t}=\epsilon_{t}+\theta_{1}\epsilon_{t-1}, \theta_{1}=0.7$

```r
MA1<-arima.sim(list(ma=0.7), n=200) 
ts.plot(MA1, main="MA(1)")
par(mfrow=c(1,2))
acf(MA1, main='ACF of MA(1)')
pacf(MA1, main='PACF of MA(1)')
```

ARMA(1,1) $Z_{t}=\phi_{1} Z_{t-1} + \epsilon_{t} + \theta_{1} \epsilon_{t-1}, \phi_{1}=0.8, \theta_{1}=0.6$

```r
x<-arima.sim(list(ar=0.8, ma= -0.6), n=200)
ts.plot(x, main='x')
par(mfrow=c(2,1))
acf(x, main='ACF of ARMA(1,1)')
pacf(x,main='PACF of ARMA(1,1)')
```
